```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#plot for exercises 1 , prob vs timestep, for various starting points, with error bars on the prob

fig1 = plt.figure()

#defines constant parameters for the plots
T=1
capsize = 3
capthick = 2

plt.xlabel('Timestep')
plt.ylabel('Probability of reaching the well')

plt.xscale('log')

#plt.title("Simple Plot")

#for X = [-1,5,1]
probA = [0.5579, 0.6129, 0.6622, 0.6802, 0.6916, 0.7049, 0.7084, 0.701, 0.71, 0.727]
conf_intA = [[0.5481644426648226, 0.5676355573351773],
  [0.6033516401714465, 0.6224483598285535],
  [0.6529285685535705, 0.6714714314464295],
  [0.671057185168488, 0.689342814831512],
  [0.6825466969615085, 0.7006533030384915],
  [0.6959593160979687, 0.7138406839020313],
  [0.6994904572357223, 0.7173095427642777],
  [0.6920253610943029, 0.709974638905697],
  [0.7011049057534715, 0.7188950942465284],
  [0.718266850202165, 0.735733149797835]]
error_A = np.transpose(np.abs(conf_intA - np.outer(probA,[1,1])) )
N = np.array([10,20,50,100,200,500,1000,2000,5000,10000])
dt = T/N
PDEProb = 0.7173050315908374

plt.errorbar(dt, probA, yerr = error_A , label='X0 = [-1.5, 1]', fmt='x-', capsize=capsize, capthick=capthick)
plt.plot(dt,[PDEProb for i in N], label= "PDE for [-1.5, 1]")

plt.legend()

plt.show()
```
